```python
import pystan
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import arviz
from sklearn import metrics
```

```python
train = pd.read_csv("../../data/kaggle-titanic/train_fixed.csv")
test = pd.read_csv("../../data/kaggle-titanic/test_fixed.csv")
```

```python
sum_pclass = train.groupby("Pclass")["Survived"].mean().sum()
mapping_pclass = train.groupby("Pclass")["Survived"].mean() / sum_pclass
train["Pclass"] = train["Pclass"].map(mapping_pclass)
test["Pclass"] = test["Pclass"].map(mapping_pclass)
```

```python
sum_embarked = train.groupby("Embarked")["Survived"].mean().sum()
mapping_embarked = train.groupby("Embarked")["Survived"].mean() / sum_embarked
train["Embarked"] = train["Embarked"].map(mapping_embarked)
test["Embarked"] = test["Embarked"].map(mapping_embarked)
```

```python
sum_title = train.groupby("Title")["Survived"].mean().sum()
mapping_title = train.groupby("Title")["Survived"].mean() / sum_title
train["Title"] = train["Title"].map(mapping_title)
test["Title"] = test["Title"].map(mapping_title)
```

```python
sum_cabin_alphabet = train.groupby("Cabin_alphabet")["Survived"].mean().sum()
mapping_cabin_alphabet = train.groupby("Cabin_alphabet")["Survived"].mean() / sum_cabin_alphabet
train["Cabin_alphabet"] = train["Cabin_alphabet"].map(mapping_cabin_alphabet)
test["Cabin_alphabet"] = test["Cabin_alphabet"].map(mapping_cabin_alphabet)
```

```python
train["Survived"] = train["Survived"].astype(int)
```

```python
data = {}
data["I"] = len(train)
data["D"] = train.shape[1]-1
data["X"] = train.drop("Survived",axis=1)
data["Survived"] = train["Survived"]
data["I_new"] = len(test)
data["X_new"] = test
```

```python
data
```

```python
fit = pystan.stan(file="model.stan"
                 ,data = data
                 ,iter=2000
                 ,chains=4
                 ,thin=1
                 ,warmup=1000
                 ,seed=1
                 )
```

```python
arviz.plot_trace(fit)
```

```python
q = pd.DataFrame(fit["q"])
q_new = pd.DataFrame(fit["q_new"])
```

```python
q_desc = q.describe().T
```

```python
q_new_desc = q_new.describe().T
```

```python
fpr, tpr, thresholds = metrics.roc_curve(train["Survived"],q_desc["50%"])

auc = metrics.auc(fpr, tpr)

plt.plot(fpr, tpr, label='ROC curve (area = %.3f)'%auc)
plt.legend()
plt.title('ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.grid(True)
```

```python
threshold = [0.5,0.6,0.7,0.8]
for value in threshold:
    print(value,":",metrics.accuracy_score(train["Survived"],[1 if x > value else 0 for x in q_desc["50%"]]))
```

```python
submission = np.array([1 if x > 0.6 else 0 for x in q_new_desc["50%"]])
```

```python
submission = pd.Series(submission)
index = pd.Series(np.arange(892,1310))
submission = pd.concat([index,submission],axis=1)
```

```python
submission.columns = ["PassengerId","Survived"]
```
